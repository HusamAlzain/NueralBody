## Import libraries and load the dataset.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic_data = pd.read_csv(url)
```

```python
# Display the first 5 rows of the dataset
titanic_data.head()
```

## Data preprocessing and cleaning.

```python
# Drop unnecessary columns
columns_to_drop = ['Name', 'Ticket', 'Cabin']
titanic_data.drop(columns=columns_to_drop, inplace=True)
```

```python
# Fill missing values in 'Age' and 'Embarked' columns
titanic_data['Age'].fillna(titanic_data['Age'].median(), inplace=True)
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True)
```

```python
# Display the cleaned dataset
titanic_data.head()
```

## Exploratory data analysis (EDA).

```python
# Visualize the distribution of passengers by gender
plt.figure(figsize=(10, 6))
sns.countplot(x='Sex', data=titanic_data)
plt.title('Passenger Distribution by Gender')
plt.show()
```

```python
# Visualize the distribution of passengers by class
plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', data=titanic_data)
plt.title('Passenger Distribution by Class')
plt.show()
```

```python
# Visualize the distribution of passengers by embarked location
plt.figure(figsize=(10, 6))
sns.countplot(x='Embarked', data=titanic_data)
plt.title('Passenger Distribution by Embarked Location')
plt.show()
```

```python
# Visualize the survival rate by gender
plt.figure(figsize=(10, 6))
sns.barplot(x='Sex', y='Survived', data=titanic_data)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
# Visualize the survival rate by class
plt.figure(figsize=(10, 6))
sns.barplot(x='Pclass', y='Survived', data=titanic_data)
plt.title('Survival Rate by Class')
plt.show()
```

```python
# Visualize the survival rate by embarked location
plt.figure(figsize=(10, 6))
sns.barplot(x='Embarked', y='Survived', data=titanic_data)
plt.title('Survival Rate by Embarked Location')
plt.show()
```

```python
# Visualize the relationship between age and survival rate
plt.figure(figsize=(10, 6))
sns.histplot(x='Age', hue='Survived', data=titanic_data, kde=True, bins=20)
plt.title('Age Distribution by Survival')
plt.show()
```

```python
# Visualize the relationship between fare and survival rate
plt.figure(figsize=(10, 6))
sns.histplot(x='Fare', hue='Survived', data=titanic_data, kde=True, bins=20)
plt.title('Fare Distribution by Survival')
plt.show()
```

```python
# Visualize the relationship between the number of siblings/spouses and survival rate
plt.figure(figsize=(10, 6))
sns.barplot(x='SibSp', y='Survived', data=titanic_data)
plt.title('Survival Rate by Number of Siblings/Spouses')
plt.show()
```

```python
# Visualize the relationship between the number of parents/children and survival rate
plt.figure(figsize=(10, 6))
sns.barplot(x='Parch', y='Survived', data=titanic_data)
plt.title('Survival Rate by Number of Parents/Children')
plt.show()
```
